```python
# drop all where status = New as these are recent ones that have not had a chance to be attended to yet
df = df[df['STATUS'] != 'New']
```

```python

# high percentage of missing values in  TOTAL_HANDLING_TIME_SECONDS and QUEUE_WAITING_TIME_SECONDS
# will investigate further
df_null_check = df.copy()
df_null_check['TOTAL_HANDLING_TIME_SECONDS_NULL'] = df_null_check['TOTAL_HANDLING_TIME_SECONDS'].isnull().astype(int)
df_null_check['QUEUE_WAITING_TIME_SECONDS_NULL'] = df_null_check['QUEUE_WAITING_TIME_SECONDS'].isnull().astype(int)
df_null_check['RESPONSE_TIME_SECONDS_NULL'] = df_null_check['RESPONSE_TIME_SECONDS'].isnull().astype(int)
grouping_counts = df_null_check.groupby(["CHANNEL", "TOTAL_HANDLING_TIME_SECONDS_NULL", "QUEUE_WAITING_TIME_SECONDS_NULL", "RESPONSE_TIME_SECONDS_NULL"]).size().reset_index(name='count').sort_values(by='CHANNEL', ascending=False)


# based off of the following emails only have response time 
# while chat and call have response,handling and waiting times



grouping_counts
```

```python
# from looking at a sample of the table it appears that QUEUE_WAITING_TIME_SECONDS = RESPONSE_TIME_SECONDS
# which is proven below 

non_null_queues = df[df['QUEUE_WAITING_TIME_SECONDS'].notnull()].copy()  # Use .copy() to avoid the warning

non_null_queues.loc[:, 'matches'] = df['QUEUE_WAITING_TIME_SECONDS'] == df['RESPONSE_TIME_SECONDS']
all_match = non_null_queues['matches'].all()

mismatch_count = (~non_null_queues['matches']).sum()
# Fill NaNs with 0
df['TOTAL_HANDLING_TIME_SECONDS'].fillna(0, inplace=True)
df['QUEUE_WAITING_TIME_SECONDS'].fillna(0, inplace=True)

print(f"There are {mismatch_count} instances where QUEUE_WAITING_TIME_SECONDS does not equal RESPONSE_TIME_SECONDS")
```

```python
df['CREATED_AT'] = pd.to_datetime(df['CREATED_AT'], errors='coerce')
df['ENDED_AT'] = df['CREATED_AT'] + pd.to_timedelta(df['TOTAL_HANDLING_TIME_SECONDS'], unit='s')  + pd.to_timedelta(df['RESPONSE_TIME_SECONDS'], unit='s')

df = df.sort_values(by=['MERCHANT_ID', 'REASON', 'DETAILED_REASON', 'CREATED_AT'])

# Calculate time difference between consecutive rows within each group (REASON, DETAILED_REASON, MERCHANT_ID)
df['TIME_DIFF'] = df.groupby(['MERCHANT_ID', 'REASON', 'DETAILED_REASON'])['CREATED_AT'].diff().dt.days

# If time difference is greater than 7 days, or it's a new group (different REASON, DETAILED_REASON, or MERCHANT_ID), mark it for a new group
df['NEW_GROUP'] = (df['TIME_DIFF'] > 7) | (df['TIME_DIFF'].isna())

# Use cumulative sum of the new group markers to assign unique ISSUE_ID
df['ISSUE_ID'] = df['NEW_GROUP'].cumsum()

# Drop temporary columns if no longer needed
df.drop(columns=['TIME_DIFF', 'NEW_GROUP'], inplace=True)
df.head()
```

```python


# Group by 'ISSUE_ID' and aggregate min 'CREATED_AT', max 'ENDED_AT', final 'STATUS', and count of touchpoints
grouped = df.groupby('ISSUE_ID').agg(
    MIN_CREATED_AT=('CREATED_AT', 'min'),
    MAX_ENDED_AT=('ENDED_AT', 'max'),
    FINAL_STATUS=('STATUS', 'last'),  # Get the last status value per ISSUE_ID
    CHANNEL_LIST=('CHANNEL', lambda x: list(x)),  # Collect all channel values into a list
    TOTAL_TOUCHPOINTS=('ISSUE_ID', 'size')  # Count the number of rows (touchpoints) for each ISSUE_ID
).reset_index()

# Merge the grouped results back into the original DataFrame on 'ISSUE_ID'
df = df.merge(grouped[['ISSUE_ID', 'MIN_CREATED_AT', 'MAX_ENDED_AT', 'FINAL_STATUS', 'CHANNEL_LIST', 'TOTAL_TOUCHPOINTS']], 
              on='ISSUE_ID', how='left')

# Calculate the total resolution time in seconds and days
df['TOTAL_RESOLUTION_TIME_SECONDS'] = (df['MAX_ENDED_AT'] - df['MIN_CREATED_AT']).dt.total_seconds()
df['TOTAL_RESOLUTION_TIME_MINUTES'] = df['TOTAL_RESOLUTION_TIME_SECONDS'] / 60  # Convert seconds to minutes
df['TOTAL_RESOLUTION_TIME_DAYS'] = df['TOTAL_RESOLUTION_TIME_SECONDS'] / 86400  # Convert seconds to days

# Create the FINAL_CHANNEL column based on the specified conditions in 'CHANNEL_LIST'
def determine_final_channel(channel_list):
    if 'email' in channel_list:
        return 'email'
    elif 'chat' in channel_list:
        return 'chat'
    elif 'call' in channel_list:
        return 'call'
    else:
        return 'unknown'  # Default value if no channel is found

df['FINAL_CHANNEL'] = df['CHANNEL_LIST'].apply(determine_final_channel)
#df = df[~((df['FINAL_CHANNEL'].isin(['chat', 'call'])) & (df['TOTAL_TOUCHPOINTS'] > 1))]


df.drop(columns=['CHANNEL_LIST'], inplace=True)

df.head()
```

```python
priority_mapping = {
    'Blocked/defective readers': 1,
    'Outage': 1,
    'Security': 1,
    
    'Connectivity issues': 2,
    'Payouts': 2,
    'Merchant verification': 2,
    'Other technical issues': 2,
    'Orders & Logistics': 2,
    'Fee negotiation': 2,
    
    'Changing account details': 3,
    'Signup': 3,
    'Bookkeeping': 3,
    'Product setup/activation': 3,
    'Other': 3,
    'Other questions about product usage': 3,
    'Profile/account/store closure': 3,
    'Pre-Signup': 3,
    'GDPR': 3,
    'No-Reply/Duplicate/SPAM': 3,
    'MISC': 3,
    'POS Pro': 3,
    'POS': 3
}

# Step 2: Apply the priority mapping to create a new 'Priority' column
df['PRIORITY'] = df['REASON_GROUP'].map(priority_mapping)
df.head()
```

```python
# Group by 'ISSUE_ID' and find the index of the row with the maximum 'CREATED_AT' for each 'ISSUE_ID'
idx = df.groupby('ISSUE_ID')['CREATED_AT'].idxmax()

# Use the index to filter the DataFrame, keeping only the row with the maximum 'CREATED_AT' per 'ISSUE_ID'
df_issue_id = df.loc[idx].reset_index(drop=True)
df_issue_id.drop(columns=['TOUCHPOINT_ID', 'CHANNEL', 'STATUS', 'TOTAL_HANDLING_TIME_SECONDS', 'RESPONSE_TIME_SECONDS',
                         'CREATED_AT', 'ENDED_AT'], inplace=True)

# Display the resulting DataFrame with one row per 'ISSUE_ID'
df_issue_id.head()
```

```python


# Define a function to detect and remove outliers based on Z-score for a given column
def detect_outliers(df, channel, time_column, z_threshold=3.5):
    # Filter data by channel
    channel_df = df[df['FINAL_CHANNEL'] == channel].copy()
    
    # Compute Z-scores for the specified time column
    z_scores = stats.zscore(channel_df[time_column])
    
    # Add Z-scores to the DataFrame
    channel_df['Z_SCORE'] = z_scores
    
    # Identify outliers where absolute Z-score is greater than the threshold
    outliers = channel_df[channel_df['Z_SCORE'].abs() > z_threshold]
    
    # Calculate percentage of outliers
    outlier_percentage = (len(outliers) / len(channel_df)) * 100 if len(channel_df) > 0 else 0
    
    # Create a DataFrame of non-outliers
    non_outliers = channel_df[channel_df['Z_SCORE'].abs() <= z_threshold]
    
    # Print out the percentage of outliers
    print(f"Percentage of outliers dropped for '{channel}': {outlier_percentage:.2f}%")
    
    return non_outliers, outliers

# Assuming df_group_id is your DataFrame with the relevant columns
# Outlier detection for 'call' channel using 'TOTAL_RESOLUTION_TIME_SECONDS'
call_non_outliers, call_outliers = detect_outliers(df_issue_id, 'call', 'TOTAL_RESOLUTION_TIME_SECONDS')

# Outlier detection for 'chat' channel using 'TOTAL_RESOLUTION_TIME_SECONDS'
chat_non_outliers, chat_outliers = detect_outliers(df_issue_id, 'chat', 'TOTAL_RESOLUTION_TIME_SECONDS')

# Outlier detection for 'email' channel using 'TOTAL_RESOLUTION_TIME_DAYS'
email_non_outliers, email_outliers = detect_outliers(df_issue_id, 'email', 'TOTAL_RESOLUTION_TIME_DAYS')

# Optionally, you can also concatenate all non-outliers into one DataFrame if needed
all_non_outliers = pd.concat([call_non_outliers, chat_non_outliers, email_non_outliers], ignore_index=True)

# Display the final DataFrames if necessary
print("Non-Outliers DataFrame:")
all_non_outliers.head()
```
